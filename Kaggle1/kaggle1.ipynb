## apply pre emphasis filter

```python
alpha = 0.9
def apply_pre_emphasis(time_series, alpha):
    # Initialize an array for the filtered time series
    filtered_series = np.zeros_like(time_series)

    # Apply the pre-emphasis filter
    for i in range(1, len(time_series)):
        filtered_series[i] = time_series[i] - alpha * time_series[i - 1]

    return filtered_series





```

```python
def denoise_column(column):
    std_dev = np.std(column)
    threshold = 6 * std_dev
    # Apply FFT to the column
    column_freq_domain = fft(column)
    
    # Calculate the magnitude spectrum
    magnitude_spectrum = np.abs(column_freq_domain)
    
    # Identify noise frequencies using a threshold
    noise_indices = np.where(magnitude_spectrum < threshold)
    
    # Set the magnitudes at noise frequencies to zero
    column_freq_domain[noise_indices] = 0
    
    # Apply IFFT to get the denoised column
    denoised_column = ifft(column_freq_domain)
    #remove complex part
    denoised_column = np.real(denoised_column)
    return denoised_column
```
